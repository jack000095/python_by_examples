### 製作 requirements 的技巧 ###
#### 方法一: 自動法

``` bash
$ pip3 freeze 

$ pip3 freeze > requirements.txt
```

方便但是有缺點:
1. 不容易升級使用中的 module  (==)
2. 因為相依性，不容易看出主要套件

### 虛擬環境的操作 ###

1.建立虛擬環境
  - pip install virtualenv
  - virtualenv env
  or only in Python 3.x 
  - python -m venv env


2.啟動虛擬環境 (site演示)
  - source env/bin/activate (linux-like/Mac OS)
  - env\Scripts\activate (Windows)

3.進行相關作業
  - pip3 ....
  
4.解除虛擬環境
  - deactivate

## Feature branch 管理 ##

### git flow feature start ###

``` bash
$ git flow feature start new-feature
Switched to a new branch 'feature/new-feature'

Summary of actions:
- A new branch 'feature/new-feature' was created, based on 'develop'
- You are now on branch 'feature/new-feature'
```

### git flow feature finish ###

``` bash
$ git flow feature finish new-feature
Switched to branch 'develop'
Updating 3145887..166db0d
Fast-forward
    new-tech.py | 0
    1 file changed, 0 insertions(+), 0 deletions(-)
    create mode 100644 new-tech.py
Deleted branch feature/new-feature (was 166db0d).
```
- code is merged to develop branch and wait for testing
- feature branch will be deleted

## Release branch 管理 ##

### git flow release start ###

``` bash
$ git flow release start 1.1.0
Switched to a new branch 'release/1.1.0'
```

### git flow release finish ###
``` bash
$ git flow release finish 1.1.0
```
- pull remote and update
- code is merged to develop and master branches
- will be tagged as 1.1.0
- branch will be deleted
- next action start with develop branch

## Hotfix branch 管理 ##

### git flow hotfix start ###
``` bash
$ git flow hotfix start quick-fix-001
```

branch hotfix/quick-fix-001 based on **master** will be created

### git flow hotfix finish ###
``` bash
$ git flow hotfix finish quick-fix-001
```
- code is merged to master and develop branches
- will be tagged
- branch will be deleted
- next action start with develop branch
